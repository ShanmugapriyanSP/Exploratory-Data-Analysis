```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('.\\files\\2018-2010_import.csv')
data_export = pd.read_csv('.\\files\\2018-2010_export.csv')
dataset.head()
dataset.shape
data_export.shape
```

```python
dataset.sort_values('value', ascending = False)
dataset.head()
```

```python
dataset.isnull().sum()
dataset.info()
#print(dataset['value'] == 'None')
```

```python
sns.heatmap(dataset.isnull(), yticklabels=False)
```

```python
from sklearn.impute import SimpleImputer
#help(SimpleImputer)
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer = imputer.fit(dataset.iloc[:,2:3])
dataset.iloc[:,2:3] = imputer.transform(dataset.iloc[:,2:3])
```

```python
df_country = dataset.groupby('country').agg({'value':'sum'})
df_country = df_country.sort_values('value', ascending=False)
df_country = df_country.iloc[:10]
print(df_country.head())
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(df_country.value, df_country.index)
plt.show()
```

```python
df_import = dataset.groupby('year').agg({'value':'sum'})
df_import = df_import.sort_values('year')
print(df_import.head())

df_export = data_export.groupby('year').agg({'value':'sum'})
df_export = df_export.sort_values('year')
print(df_export.head())

plt.style.use('classic')
plt.plot(df_import.index, df_import.value, label='Import')
plt.plot(df_export.index, df_export.value, label='Export')
plt.legend(loc='best')
```
